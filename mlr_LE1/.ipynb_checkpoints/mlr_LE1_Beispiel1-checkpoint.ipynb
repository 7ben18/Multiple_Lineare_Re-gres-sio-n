```python
# Module importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Normalengleichung
# Ax = y

# Features und Target als np.array definieren
x1 = np.array([1,2,3,4,1,2,3,4]).reshape(-1,1) # Kinder
x2 = np.array([0,1,2,3,2,3,4,5]).reshape(-1,1) # Katzen
x3 = np.ones((8,1)) # Arrays mit 1 erstellen

# Target Array
y = np.array([2,3,4,5,0,1,2,3]).reshape(-1,1)

# Konkardinieren von x1 und x2 np.array zu einer 8 x 2 Matrix
A = np.hstack([x1, x2, x3])

# Ausgabe von beiden Matrizen
display(A, y)
```

```python
# Normalengleichung nach x aufloesen
x = np.linalg.inv(A.T @ A) @ A.T @ y 
display(x)
```

```python
b0 = x[-1]
b1 = x[0]
b2 = x[1]
```

```python
data = {"Kinder": [1,2,3,4,1,2,3,4],
     "Autos": [0,1,2,3,2,3,4,5],
     "Katzen": [2,3,4,5,0,1,2,3]}
df = pd.DataFrame(data)
df
```

```python
df[["Katzen", "Autos"]]
```

```python
plt.scatter(x = df["Autos"], y = df["Katzen"], color = "blue")
plt.xlabel("Katzen", fontsize = 15)
plt.ylabel("Autos", fontsize = 15)
plt.title("Autos vs Katzen", fontsize = 15)
plt.grid()
plt.show()
```

```python
# Sklearn Modul importieren
from sklearn.linear_model import LinearRegression

# X und y definieren
x = df["Autos"].values.reshape(-1,1)
y = df["Katzen"]

# Model erstellen
model = LinearRegression().fit(x, y)

# Steigung und Achsenabschnitt
steigung = model.coef_
achsenabschnitt = model.intercept_

# X Werte generieren
x_values = np.linspace(0,max(df["Autos"]),5)

# y_predict Werte berechnen aus generierten X Werten
y_predict = x_values * steigung + achsenabschnitt

# Visualisieren
plt.scatter(x = df["Autos"], y = df["Katzen"], color = "blue")
plt.plot(x_values, y_predict, color = "red")
plt.xlabel("Autos", fontsize = 15)
plt.ylabel("Katzen", fontsize = 15)
plt.title("Katzen vs Autos", fontsize = 15)
plt.grid()
plt.show()

# Interpretation 
# Univariate Lineare Regression funktioniert fuer Katzen nicht. 
# Je mehr Autos desto mehr Katzen kann nicht sein. 
```
